```python
from tkinter import *
from tkinter import messagebox

root = Tk()
root.geometry("400x250")
root.configure(bg="#E8F8F5")


def clear_entries():
    entName.delete(0, END)
    entSurname.delete(0, END)
    entId.delete(0, END)
    entSearch.delete(0, END)
    entNewName.delete(0, END)
    entNewSurname.delete(0, END)
    entNewID.delete(0, END)
    entNewInfo.delete(0, END)
    varVis.set(0)
    varOp.set(0)

def show_add_page():
    main_page.pack_forget()
    add_page.pack()
    
def show_search_page():
    main_page.pack_forget()
    search_page.pack()
    
def show_edit_page():
    info_page.pack_forget()
    edit_page.pack()
    
def show_main_page():
    add_page.pack_forget()
    search_page.pack_forget()
    edit_page.pack_forget()
    info_page.pack_forget()
    main_page.pack()
    clear_entries()
    
def show_info_page(patientInfo):
    search_page.pack_forget()
    info_page.pack()
    
    text_info.delete(1.0, END)
    
    with open('d:/patientlist.txt','r') as file:
        for line in file:
            if patientInfo in line:
                text_info.insert(END, line)
                break

def save():
    name=entName.get()
    surname=entSurname.get()
    Id=entId.get()
    visit = varVis.get()
    operation = varOp.get()
    
    if visit and operation:
        info = "Visit and Operation"
    elif visit:
        info = "Visit"
    elif operation:
        info = "Operation"
    
    f = open('d:/patientlist.txt','a')
    f.write(name+' '+surname+' '+"Id: "+Id+' , '+info+'\n')
    f.close()
    messagebox.showinfo(message='Patient Added Successfully')
    add_page.pack_forget()
    main_page.pack()
    clear_entries()
    
def search():
    searchResult = entSearch.get()
    found = False
    
    with open('d:/patientlist.txt','r') as file:
        for line in file:
            if searchResult in line:
                found = True
                break   
                
    if found:
        show_info_page(searchResult)
    else:
        messagebox.showinfo(message='Patinet Not Found!',title='Error')
               
def edit():
    newName = entNewName.get()
    newSurname = entNewSurname.get()
    newID = entNewID.get()
    newInfo = entNewInfo.get()

    editedPatientInfo = f"{newName} {newSurname} Id: {newID}, {newInfo}\n"
    
    with open('d:/patientlist.txt', 'r') as file:
        lines = file.readlines()

    with open('d:/patientlist.txt', 'w') as file:
        
        for line in lines:
            if f"Id: {newID}" not in line:
                file.write(line)
                
        file.write(editedPatientInfo)
    
    edit_page.pack_forget()
    main_page.pack()
        
def on_enter(e):
    e.widget['background'] = '#E67E22'

def on_leave(e):
    e.widget['background'] = '#16A085'        
    
root.title("Doctor Assistant")    
        
#Main Page 
main_page = Frame(root, bg='#E8F8F5')
lblWelcome = Label(main_page, text='Welcome to Doctor Assisstant!'+'\n'+'Please Choose Your Action:',fg='#16A085',bg='#E8F8F5',font=('Segoe UI',16)) 
lblWelcome.grid(row=0,column=0,columnspan=2, pady=40)

btnAdd = Button(main_page, text='Add New Patient',command=show_add_page, fg='#E8F8F5', bg='#16A085')
btnAdd.grid(row=1,column=0,padx=50)
btnAdd.bind("<Enter>", on_enter)
btnAdd.bind("<Leave>", on_leave)

btnSearch = Button(main_page, text='Search Patient',command=show_search_page, fg='#E8F8F5',bg='#16A085')
btnSearch.grid(row=1,column=1,padx=50)
btnSearch.bind("<Enter>", on_enter)
btnSearch.bind("<Leave>", on_leave)

#Add Page
add_page = Frame(root, bg='#E8F8F5')
lblName = Label(add_page, text='Name: ',bg='#E8F8F5')
lblName.grid(row=0, column=0, pady=10)
entName = Entry(add_page)
entName.grid(row=0, column=1, pady=10)
lblSurname = Label(add_page, text='Surname: ',bg='#E8F8F5')
lblSurname.grid(row=1, column=0, pady=11)
entSurname = Entry(add_page)
entSurname.grid(row=1, column=1, pady=11)
lblId = Label(add_page, text='ID: ',bg='#E8F8F5')
lblId.grid(row=2, column=0, pady=12)
entId = Entry(add_page)
entId.grid(row=2, column=1, pady=12)
lblInfo = Label(add_page, text='Info: ',bg='#E8F8F5')
lblInfo.grid(row=3, column=0, pady=13)
varVis = IntVar()
varOp = IntVar()
chkVis = Checkbutton(add_page, text='Visit', bg='#E8F8F5', var=varVis)
chkVis.grid(row=3, column=1, pady=13)
chkOp = Checkbutton(add_page, text='Operation', bg='#E8F8F5', var=varOp) 
chkOp.grid(row=3, column=2, pady=13)
btnSave = Button(add_page, text='Save', command=save, fg='#E8F8F5',bg='#16A085')
btnSave.grid(row=4, column=0, pady=14)
btnSave.bind("<Enter>", on_enter)
btnSave.bind("<Leave>", on_leave)
btnback = Button(add_page, text='Back', command=show_main_page, fg='#E8F8F5',bg='#16A085')    
btnback.grid(row=4, column=1, pady=14)
btnback.bind("<Enter>", on_enter)
btnback.bind("<Leave>", on_leave)

#Search Page
search_page = Frame(root, bg='#E8F8F5')
lblSearch = Label(search_page, text='Patient ID: ',bg='#E8F8F5')
lblSearch.grid(row=0, column=0, pady=30)
entSearch = Entry(search_page)
entSearch.grid(row=0, column=1, pady=30)
btnSearch = Button(search_page, text='Search', command=search, fg='#E8F8F5',bg='#16A085')
btnSearch.grid(row=1, column=0, pady=32)
btnSearch.bind("<Enter>", on_enter)
btnSearch.bind("<Leave>", on_leave)
btnback = Button(search_page, text='Back', command=show_main_page, fg='#E8F8F5',bg='#16A085')
btnback.grid(row=1, column=1, pady=32)
btnback.bind("<Enter>", on_enter)
btnback.bind("<Leave>", on_leave)

#Info Page
info_page = Frame(root, bg='#E8F8F5')
text_info = Text(info_page, width=30, height=5)
text_info.pack()
btnEdit = Button(info_page, text='Edit', command=show_edit_page, fg='#E8F8F5',bg='#16A085')
btnEdit.pack()
btnEdit.bind("<Enter>", on_enter)
btnEdit.bind("<Leave>", on_leave)
btnBACK = Button(info_page, text='Back', command=show_main_page, fg='#E8F8F5',bg='#16A085')
btnBACK.pack()
btnBACK.bind("<Enter>", on_enter)
btnBACK.bind("<Leave>", on_leave)

#Edit Page
edit_page = Frame(root, bg='#E8F8F5')
lblNewName = Label(edit_page, text='New Name: ',bg='#E8F8F5')
lblNewName.pack()
entNewName = Entry(edit_page)
entNewName.pack()    
lblNewSurname = Label(edit_page, text='New Surname: ',bg='#E8F8F5')
lblNewSurname.pack()
entNewSurname = Entry(edit_page)
entNewSurname.pack()    
lblNewID = Label(edit_page, text='New ID: ',bg='#E8F8F5')
lblNewID.pack()
entNewID = Entry(edit_page)
entNewID.pack()    
lblNewInfo = Label(edit_page, text='New Info: ',bg='#E8F8F5')
lblNewInfo.pack()
entNewInfo = Entry(edit_page)
entNewInfo.pack()
btnsave = Button(edit_page, text='Save', command=edit, fg='#E8F8F5',bg='#16A085')
btnsave.pack()
btnsave.bind("<Enter>", on_enter)
btnsave.bind("<Leave>", on_leave)
btnBack = Button(edit_page, text='Back',command=show_main_page, fg='#E8F8F5',bg='#16A085')
btnBack.pack()
btnBack.bind("<Enter>", on_enter)
btnBack.bind("<Leave>", on_leave)
main_page.pack()
    
root.mainloop()
```
